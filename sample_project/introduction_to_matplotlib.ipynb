## Matplotlib example workflow


```python
# 0. Import Matplotlib and get it ready for plotting in Jupyter
import matplotlib.pyplot as plt
%matplotlib inline

# 1. Prepare data
x = [1, 4, 2, 90, 45, 3]
y = [28, 0, 98, 555, 36, 40]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 5))  # (width,height)

# Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Sample Plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("../images/sample-plot.png")
```

## Two options for subplots


```python
# Subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
# Plot to each different axis
ax1.plot(x, x/4, 'tab:orange')
ax2.scatter(x, x**8)
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(x, bins=10, color='tab:red')
# ax4.hist(x,color='tab:red')
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
# Plot on each different index
ax[0, 0].plot(x, x/4, color='tab:orange')
ax[0, 0].set(title="Line Plot")
ax[0, 1].scatter(x, x**8)
ax[0, 1].set_title("Scatter Plot")
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 0].set(title="Bar Plot")
ax[1, 1].hist(x, color='tab:red')
ax[1, 1].set_title("Histogram")
```

### Create a scatter plot of x and y values, where x is a list of integers from 1 to 10 and y is the square of each x value.


```python
x = np.arange(1, 11)
y = x**2
fig, ax = plt.subplots()
ax.scatter(x=x, y=y)
ax.set(xlabel="x-axis", ylabel="y-axis", title="Plot Title")
```

### Create a line plot of the sine function from 0 to 2π.


```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
fig, axs = plt.subplots(nrows=2, ncols=1)
axs[0].plot(x, y1, color="blue", linewidth=2)
axs[0].set_title("Sine Function")

axs[1].plot(x, y2, color="red", linewidth=2)
axs[1].set_title("Cosine Function")

fig.suptitle("Sine and Cosine Function")
fig.subplots_adjust(hspace=0.3)
```

## Which one should I use? (Pyplot vs Matplotlib OO Method(Object Oriented Method)?)

- When plotting something quickly, okay to use the pyplot method.
- When plotting something more advanced, use the OO method.


```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
len(heart_disease["age"])
```

```python
len(over_50)
```

```python
# PyPlot Method
over_50.plot(kind="scatter", x="age", y="chol", c="target")
```

```python
# OO Method mixed with PyPlot method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax)

# ax.set_xlim(45,100);
```

```python
# OO Method from Scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])

# Customized the plot
ax.set(title="Heart Disease & Cholesterol Level",
       xlabel="Age", ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line
ax.axhline(y=over_50["chol"].mean(), linestyle='dashed')
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])
# Customize ax0
ax0.set(title="Heart Disease & Cholesterol Level", ylabel="Cholesterol")
# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# Add a meanline
ax0.axhline(y=over_50["chol"].mean(), linestyle="dashed")

# Add data to ax1
scatter1 = ax1.scatter(
    x=over_50["age"], y=over_50["thalach"], c=over_50["target"])
# Customize ax1
ax1.set(title="Heart Disease & Max Heart Rate",
        xlabel="Age", ylabel="Max Heart Rate")
# Add a legend to ax1
ax1.legend(*scatter1.legend_elements(), title="Target")
# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(), linestyle="dashed")

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
```
